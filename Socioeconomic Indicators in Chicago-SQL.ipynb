## Questions and Solutions

### Problem 1

##### How many rows are in CENSUS_DATA?

```python
df = pd.read_sql_query("SELECT COUNT(*) FROM CENSUS_DATA", conn)
df
```

### Problem 2

##### How many community areas in Chicago have a hardship index greater than 50.0?

```python
df = pd.read_sql_query("SELECT COUNT(*) FROM CENSUS_DATA WHERE HARDSHIP_INDEX > 50.0", conn)
df
```

### Problem 3

##### What is the maximum value of hardship index in CENSUS_DATA?

```python
df = pd.read_sql_query("SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA", conn)
df
```

### Problem 4

##### Which community area which has the highest hardship index?

```python
df = pd.read_sql_query("SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA)", conn)
df
```

### Problem 5

##### Which Chicago community areas have per-capita incomes greater than $60,000?

```python
df = pd.read_sql_query("SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA WHERE PER_CAPITA_INCOME > 60000", conn)
df
```

### Problem 6

##### Create a scatter plot using the variables `per_capita_income_` and `hardship_index`. Explain the correlation between the two variables.

```python
df = pd.read_sql_query("SELECT PER_CAPITA_INCOME, HARDSHIP_INDEX FROM CENSUS_DATA", conn)

import matplotlib.pyplot as plt
import seaborn as sns

df.dropna(inplace=True)

plot = sns.jointplot(x = 'per_capita_income', y = 'hardship_index', data=df)
plt.show()
```

As Per Capita Income rises as the Hardship Index decreases. We see that the points on the scatter plot are somewhat closer to a straight line in the negative direction, so we have a negative correlation between the two variables. 

### Problem 7

##### How many Elementary Schools are in CHICAGO_PUBLIC_SCHOOLS?

```python
df = pd.read_sql_query("SELECT COUNT(*) FROM CHICAGO_PUBLIC_SCHOOLS WHERE Elementary_Middle_or_High_School = 'ES'", conn)
df.head()
```

### Problem 8

##### What is the highest Safety Score?

```python
df = pd.read_sql_query("SELECT MAX(SAFETY_SCORE) AS MAX_SAFETY_SCORE FROM CHICAGO_PUBLIC_SCHOOLS", conn)
df
```

### Problem 9

##### Which schools have highest Safety Score?

```python
df = pd.read_sql_query("""SELECT NAME_OF_SCHOOL, SAFETY_SCORE FROM CHICAGO_PUBLIC_SCHOOLS 
                        WHERE SAFETY_SCORE = (SELECT MAX(SAFETY_SCORE) FROM CHICAGO_PUBLIC_SCHOOLS)""", conn)
df
```

### Problem 10

##### What are the top 10 schools with the highest "Average Student Attendance"?

```python
df = pd.read_sql_query("""SELECT TOP 10 NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM CHICAGO_PUBLIC_SCHOOLS 
                        ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC""", conn)
df
```

### Problem 11

##### Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance

```python
df = pd.read_sql_query("""SELECT TOP 5 NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM CHICAGO_PUBLIC_SCHOOLS 
                        ORDER BY AVERAGE_STUDENT_ATTENDANCE""", conn)
df
```

### Problem 12

##### Now remove the '%' sign from the above result set for Average Student Attendance column

```python
df = pd.read_sql_query("""SELECT TOP 5 NAME_OF_SCHOOL, REPLACE(AVERAGE_STUDENT_ATTENDANCE, '%','') AS AVERAGE_STUDENT_ATTENDANCE
                            FROM CHICAGO_PUBLIC_SCHOOLS 
                            ORDER BY AVERAGE_STUDENT_ATTENDANCE""", conn)
df
```

### Problem 13

##### Which Schools have Average Student Attendance lower than 70%?

```python
df = pd.read_sql_query("""SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM CHICAGO_PUBLIC_SCHOOLS 
                        WHERE AVERAGE_STUDENT_ATTENDANCE<'70' 
                        ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC""", conn)
df
```

### Problem 14

##### Get the total College Enrollment for each Community Area

```python
df = pd.read_sql_query("""SELECT COMMUNITY_AREA_NAME, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT 
                       FROM CHICAGO_PUBLIC_SCHOOLS 
                       GROUP BY COMMUNITY_AREA_NAME""", conn)
df
```

### Problem 15

##### Get the 5 Community Areas with the least total College Enrollment  sorted in ascending order

```python
df = pd.read_sql_query("""SELECT TOP 5 COMMUNITY_AREA_NAME, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT 
                       FROM CHICAGO_PUBLIC_SCHOOLS 
                       GROUP BY COMMUNITY_AREA_NAME 
                       ORDER BY TOTAL_ENROLLMENT
                       """, conn)
df
```

### Problem 16

##### List 5 schools with lowest safety score.

```python
df = pd.read_sql_query("""SELECT TOP 5 NAME_OF_SCHOOL, SAFETY_SCORE FROM CHICAGO_PUBLIC_SCHOOLS 
                        ORDER BY SAFETY_SCORE""", conn)
df
```

### Problem 17

##### Get the hardship index for the community area which has College Enrollment of 4368

```python
df = pd.read_sql_query("""SELECT CPS.COMMUNITY_AREA_NAME, CD.HARDSHIP_INDEX, CPS.COLLEGE_ENROLLMENT
                       FROM CHICAGO_PUBLIC_SCHOOLS CPS, CENSUS_DATA CD
                       WHERE CD.COMMUNITY_AREA_NUMBER = CPS.COMMUNITY_AREA_NUMBER 
                       AND CPS.COLLEGE_ENROLLMENT = 4368
                       """, conn)
df
```

### Problem 18

##### Get the hardship index for the community area which has the school with the  highest enrollment.

```python
df = pd.read_sql_query("""SELECT CD.COMMUNITY_AREA_NUMBER, CPS.COMMUNITY_AREA_NAME, CD.HARDSHIP_INDEX
                       FROM CHICAGO_PUBLIC_SCHOOLS CPS, CENSUS_DATA CD
                       WHERE CPS.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER 
                       AND CPS.COLLEGE_ENROLLMENT =
                       (SELECT MAX(COLLEGE_ENROLLMENT) FROM CHICAGO_PUBLIC_SCHOOLS)
                       """, conn)
df
```

### Problem 19

### Problem 20

### Problem 21

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)

```python
df = pd.read_sql_query("SELECT CASE_NUMBER FROM CHICAGO_CRIME_DATA WHERE DESCRIPTION LIKE '%MINOR%'", conn)
df
```

### Problem 22

##### List all kidnapping crimes involving a child?

```python
df = pd.read_sql_query("""SELECT * FROM CHICAGO_CRIME_DATA 
                        WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%'
                       """, conn)
df
```

### Problem 23

##### What kinds of crimes were recorded at schools?

```python
df = pd.read_sql_query("""SELECT DISTINCT PRIMARY_TYPE FROM CHICAGO_CRIME_DATA 
                        WHERE LOCATION_DESCRIPTION LIKE '%school%'
                       """, conn)
df
```

### Problem 24

##### List the average safety score for all types of schools.


```python
df = pd.read_sql_query("""SELECT Elementary_Middle_or_High_School, AVG(CAST(SAFETY_SCORE AS INT)) AS average_safety_score
                       FROM CHICAGO_PUBLIC_SCHOOLS GROUP BY Elementary_Middle_or_High_School
                       """, conn)
df
```

### Problem 25

##### List 5 community areas with highest % of households below poverty line

```python
df = pd.read_sql_query("""SELECT TOP 5 COMMUNITY_AREA_NAME 
                       FROM CENSUS_DATA ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
                       """, conn)
df
```

### Problem 26

##### Which community area is most crime prone?

```python
df = pd.read_sql_query("""SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA
                        WHERE COMMUNITY_AREA_NUMBER = 
                        (SELECT TOP 1 COMMUNITY_AREA_NUMBER 
                        FROM CHICAGO_CRIME_DATA
                        GROUP BY COMMUNITY_AREA_NUMBER 
                        ORDER BY COUNT(*) DESC)
                       """, conn)
df
```

### Problem 27

##### Use a sub-query to find the name of the community area with highest hardship index

```python
df = pd.read_sql_query("""SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA
                        WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA)
                        """, conn)
df
```

### Problem 28

##### Use a sub-query to determine the Community Area Name with most number of crimes?

```python
df = pd.read_sql_query("""SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA
                        WHERE COMMUNITY_AREA_NUMBER = (SELECT TOP 1COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA
                        GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC)""", conn)
df
```
